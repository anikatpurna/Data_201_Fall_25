```python
#Describe what you see:
print(DF.columns) #lists all column names
print(DF.shape) #tells how big your dataset is
print(DF.dtypes) #tells what kind of data each column holds
DF.describe() #gives a statistical summary of all numeric columns
```

```python
# 2 What are the three most common first names of purchasers?
DF['first_name'].value_counts()
```

we can see that the three most common first names are Jackson, Joseph, and Jacob.

```python
# 3 What are the three most common themes of Lego sets purchased?
DF['theme'].value_counts()
```

In this sample, the three most common Lego themes purchased are Star Wars, Nexo Knights, and tied for third is Gear and Mixels.

```python
#4 Among the most common theme of Lego sets purchased, what is the most common subtheme?
common_theme="Star Wars"
mask=DF['theme']==common_theme
DF_common = DF[mask].copy()
DF_common['subtheme'].value_counts()
```

In this sample, the most common subtheme is The Force Awakens.

```python
#from professor's notebook

#5 Create data frames for each of the ages in the following categories: 
# "18 and under", "19 - 25", "26 - 35", "36 - 50", "51 and over". HINT - use masks

# create 5 masks
mask_u18 = (DF['age'] <= 18)
mask_19_25 = (DF['age'] >= 19) & (DF['age'] <=25)
mask_26_35 = (DF['age'] >= 26) & (DF['age'] <=35)
mask_36_50 = (DF['age'] >= 36) & (DF['age'] <=50)
mask_50u = (DF['age'] >= 50)

# create 5 data frames
DF_u18 = DF[mask_u18].copy()
DF_19_25 = DF[mask_19_25].copy()
DF_26_35 = DF[mask_26_35].copy()
DF_36_50 = DF[mask_36_50].copy()
DF_50u = DF[mask_50u].copy()

```

```python
#6 Which age group has purchased the highest number of Lego sets.

print('under 18:')
print(DF_u18['quantity'].sum())
print('-----')

print('19-25')
print(DF_19_25['quantity'].sum())
print('-----')

print('26-35')
print(DF_26_35['quantity'].sum())
print('-----')

print('36-50')
print(DF_36_50['quantity'].sum())
print('-----')

print('over 50')
print(DF_50u['quantity'].sum())
```

In this sample, the age group 36 - 50 spent the most money on Legos.

```python
#from professor's notebook

#7 Which age group has spent the most money on Legos?

# Add a column for money spent.
DF['money_spent'] = DF['us_price']*DF['quantity']

# Then redo the masks
DF_u18 = DF[mask_u18].copy()
DF_19_25 = DF[mask_19_25].copy()
DF_26_35 = DF[mask_26_35].copy()
DF_36_50 = DF[mask_36_50].copy()
DF_50u = DF[mask_50u].copy()

# Then sum the money spent
print('under 18:')
print(DF_u18['money_spent'].sum())
print('-----')

print('19-25')
print(DF_19_25['money_spent'].sum())
print('-----')

print('26-35')
print(DF_26_35['money_spent'].sum())
print('-----')

print('36-50')
print(DF_36_50['money_spent'].sum())
print('-----')

print('over 50')
print(DF_50u['money_spent'].sum())
```

In this sample, the age group 36 - 50 purchased the highest number of Lego sets.

```python
#8 Which Lego theme has made the most money for Lego?
colummns = ['theme','money_spent']
DF[colummns].groupby('theme').sum().sort_values('money_spent',ascending=False)
```

The Star Wars theme has made the most money for lego in this data.

```python
#9 Which area code has spent the most money on Legos? 
# In the US the area code is the first 3 digits of a phone number.

DF['area_code'] = DF['phone_number'].apply(lambda x: str(x).split('-')[0])

columns = ['area_code','money_spent']
DF[columns].groupby('area_code').sum().sort_values('money_spent',ascending=False)
```

In this sample, looks like unknown regions purchased the most but the area code 956 has spent the most money on Legos.

```python
counts = DF['area_code'].value_counts()
counts
```

```python
spent = DF[columns].groupby('area_code').sum().sort_values('money_spent',ascending=False)

average_spent = []
for c in counts.items():
    average_spent.append(spent.loc[c[0]]['money_spent']/c[1])
    # if c[0] != 'nan':
    #     print(f'Area code {c[0]} spent an average of {spent.loc[c[0]]['money_spent']/c[1]}')

spent['average_spent'] = average_spent
spent.sort_values('average_spent',ascending=False)
```

First we calculate how much money each purchase is worth (us_price * quantity).

Then we group by area code and total up the money.

We count how many purchases came from each area code.

Using a for loop, we divide the total money by the number of purchases → this gives us the average spent per purchase in that area code.

Finally, we sort the table so the area code with the highest average spending shows up at the top.
