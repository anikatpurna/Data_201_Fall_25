## Goal of This Notebook
The goal is to analyze **which countries had the fastest GDP growth between 1960 and 2020**.  

The process was:  
1. From the original dataset, I selected GDP values for **1960** and **2020**.  
2. I calculated **Growth % = ((GDP_2020 - GDP_1960) / GDP_1960) * 100**.  
3. I sorted countries by growth percentage and saved the **Top 10 fastest-growing countries** to a subset file.  
4. This notebook reads that subset (`fastest_gdp_growth_1960_2020.csv`) and visualizes the results.  

The data is saved in the file:  
**`fastest_gdp_growth_1960_2020.csv`**


```python
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
import kagglehub

pio.renderers.defaule = 'colab'
!conda install -y lxml beautifulsoup4 html5lib
!conda install -y openpyxl xlrd
!conda install -y requests
```

```python
# Read the saved subset file
df = pd.read_csv("fastest_gdp_growth_1960_2020.csv")


print("Shape:", df.shape)
```

```python
# show the data
df.head()
```

```python
## Top 10 Countries by Growth %
# Sort just in case
df_sorted = df.sort_values(by="Growth %", ascending=False)

# Display
df_sorted
```

```python
## Visualization
plt.figure(figsize=(12,6))
plt.bar(df_sorted["Country Name"], df_sorted["Growth %"])
plt.title("Top 10 Countries by GDP Growth % (1960-2020)")
plt.ylabel("Growth %")
plt.xticks(rotation=45, ha="right")
plt.show()
```

## Results & Interpretation
The analysis shows that the countries with the **fastest GDP growth** from 1960-2020 are:  

- Botswana  
- Singapore  
- Korea, Rep. (South Korea)  
- Hong Kong SAR (China)  
- China  
- Ireland  
- Thailand  
- Malaysia  

These results highlight the success stories of **industrialization and globalization** in Asia and Africa.  

Developed countries like the United States are not in the top 10 because they already had high GDP in 1960, so their relative growth percentage is smaller.  

**Conclusion:** Rapidly developing economies grew at extraordinary rates, transforming from small economies in 1960 to global players by 2020.  


## Reflection

One thing that surprised me in this analysis was seeing **Botswana** at the very top of the list.  
When people think about GDP growth, they usually focus on countries like **China** or **India**, but Botswana’s growth stood out because it started with such a small economy in 1960 and then expanded rapidly due to natural resources and stable governance.  

It was also interesting to see how **small economies like Singapore and Hong Kong** could achieve such massive growth percentages by transforming into global trade and financial hubs.  
Meanwhile, countries like the **United States or Germany** did not appear in the top 10. This makes sense because they already had very high GDP in 1960, so their percentage growth looks smaller even though their economies are still among the largest in absolute terms.  

This reflection shows that **percentage growth highlights a different story** than just looking at absolute GDP — smaller, emerging economies can shine when we measure growth rates over long periods.
