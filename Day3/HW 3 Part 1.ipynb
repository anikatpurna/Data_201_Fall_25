```python
#5. Most Kaggle datasets contain some basic stats or visualizations on the download page.
#See if you can recreate some of the plots or data you see on the website.

# dataframe for consistency
DF = df.copy()

# Example 1: Histogram of GDP in 1960 (like Kaggle mini-chart)
plt.figure(figsize=(8,6))
plt.hist(DF['1960'].dropna(), bins=30, color="skyblue", edgecolor="black")
plt.title("Distribution of GDP across Countries in 1960")
plt.xlabel("GDP in US$")
plt.ylabel("Number of Countries")
plt.show()

# Example 2: Top 20 countries by GDP in 2020 (as per your example code)
year = '2020'
top20_2020 = DF[['Country Name', year]].sort_values(by=year, ascending=False).head(20)

plt.figure(figsize=(15,8))
plt.bar(top20_2020['Country Name'], top20_2020[year], color="orange")
plt.title("Top 20 Countries by GDP in 2020")
plt.ylabel("Gross Domestic Product (US$)")
plt.xticks(rotation=90)
plt.show()
```

The histogram recreates Kaggle’s automatic summary chart, showing that most countries in 1960 had very small GDP values, while a few (like the USA) had extremely high GDP.
The bar chart for 2020 shows that the United States and China dominate, followed by other advanced economies such as Japan, Germany, and India.

```python
#6 
#My Question: “Which countries had the fastest GDP growth from 1960 to 2020?”


#Calculate GDP growth between 1960 and 2020 for each country

#Keep only relevant columns
growth_df = DF[['Country Name', '1960', '2020']].dropna()

#Calculate growth percentage
growth_df['Growth %'] = ((growth_df['2020'] - growth_df['1960']) / growth_df['1960']) * 100

#Sort countries by highest growth
fastest_growth = growth_df.sort_values(by='Growth %', ascending=False).head(10)

print("Top 10 Countries with Fastest GDP Growth (1960–2020):")
print(fastest_growth[['Country Name', 'Growth %']])
```

“Which countries had the fastest GDP growth from 1960 to 2020?”
Based on the dataset, the top 10 fastest-growing economies are:
Botswana
Singapore

East Asia & Pacific (IDA & IBRD countries)
East Asia & Pacific (excluding high income)
Thailand
Malaysia
These countries/regions had extraordinary growth rates, with Botswana, Singapore, and South Korea at the very top.
This makes sense historically:
Botswana grew rapidly after independence, largely due to diamond mining.
Singapore, South Korea, Hong Kong, and China became global economic powerhouses through industrialization and trade.
Ireland transformed its economy through technology and finance.
Thailand and Malaysia experienced strong industrial and export-led growth.
By contrast, developed economies like the United States or Germany already had high GDP in 1960, so their percentage growth looks smaller, even if their GDP levels remain very large.
